```python
import os
import shutil

# 1. Define the source folder for the bioclimatic data
bioclim_folder = "/mnt/bweeding_workspace/ACIAR/bioclim_rasters/ba_cordex_month_+xx_q100/"

# 2. Define the models to be processed (add all your model prefixes to this list)
models = [
    # "CERFACS-CNRM-CM5_SMHI-RCA4_",
    "CNRM-CERFACS-CNRM-CM5_SMHI-RCA4_"
    # Add other models here, e.g., "MPI-M-MPI-ESM-LR_SMHI-RCA4_", etc.
]

# 3. Define a map for project keywords to their full names in the filenames
project_map = {
    'hist': 'historical',
    'future': 'rcp85'
}

# 4. Map the project keywords to their destination folder names
destination_map = {
    'hist': '/scratch/gito_aciar/data/worldclim_train',
    'future': '/scratch/gito_aciar/data/worldclim_train_future'
}

# 5. Define the range of bioclimatic variables (bio 1 to 19)
bio_variables = range(1, 20)
```

```python


# --- Main Processing Loop ---
# This script copies files to the main destination folder and renames them.
# WARNING: This will overwrite files if more than one model is processed.

print("Starting file copy and rename process...")
print("!!! WARNING: Files from previous models in the list will be overwritten. !!!")
print("-" * 40)

file_copy_count = 0
file_missing_count = 0

# Loop through each model in the list
for model_prefix in models:
    print(f"\nProcessing Model: {model_prefix.strip('_')}")

    # Loop through each project type ('hist', 'future')
    for project_key, project_name in project_map.items():
        
        # Get the final destination folder
        destination_folder = destination_map[project_key]
        print(f"  -> Project: {project_name} | Destination: {destination_folder}")
        
        # Create the destination folder if it doesn't exist
        os.makedirs(destination_folder, exist_ok=True)

        # Loop through each bioclimatic variable (1-19)
        for i in bio_variables:
            
            # --- Construct the SOURCE file path ---
            source_filename = f"{model_prefix}{project_name}_ba_month_x++_q100_bio_{i}.tif"
            source_path = os.path.join(bioclim_folder, source_filename)
            
            # --- Construct the DESTINATION file path with the NEW name ---
            destination_filename = f"bio_{i}.tif"
            destination_path = os.path.join(destination_folder, destination_filename)
            
            # Check if the source file exists
            if os.path.exists(source_path):
                try:
                    # Copy source to destination, overwriting if it exists
                    shutil.copy2(source_path, destination_path)
                    print(f"    - COPIED: {source_filename}  ->  TO: {destination_path}")
                    file_copy_count += 1
                except Exception as e:
                    print(f"    - ERROR copying {source_filename}: {e}")
            else:
                # Print a warning if the source file is not found
                print(f"    - SKIPPED (Not Found): {source_filename}")
                file_missing_count += 1

print("-" * 40)
print("\nProcessing complete.")
print(f"Summary: {file_copy_count} files copied, {file_missing_count} files not found.")
```

```python
import rasterio
from rasterio.warp import reproject, Resampling
import matplotlib.pyplot as plt
import numpy as np

# Path file
# file1 = "/mnt/bweeding_workspace/ACIAR/bioclim_rasters/ba_cordex_month_x++_q100/CNRM-CERFACS-CNRM-CM5_SMHI-RCA4_historical_ba_month_x++_q100_bio_1.tif"
# file2 = "/mnt/bweeding_workspace/ACIAR/bioclim_rasters/ba_cordex_month_x++_q100/CNRM-CERFACS-CNRM-CM5_SMHI-RCA4_rcp85_ba_month_x++_q100_bio_1.tif"

file1 = "/mnt/bweeding_workspace/ACIAR/bioclim_rasters/ensemble_means/ensemble_mean_historical_ba_month_x++_q100_bio_1.tif"
file2 = "/mnt/bweeding_workspace/ACIAR/bioclim_rasters/ensemble_means/ensemble_mean_rcp85_ba_month_x++_q100_bio_1.tif"


# Buka file pertama
with rasterio.open(file1) as src1:
    data1 = src1.read(1, masked=True)
    profile1 = src1.profile

# Buka file kedua
with rasterio.open(file2) as src2:
    data2 = src2.read(1, masked=True)

    # Reproject jika kedua file punya CRS
    if src1.crs is not None and src2.crs is not None:
        data2_resampled = np.empty_like(data1, dtype=np.float32)

        reproject(
            source=data2,
            destination=data2_resampled,
            src_transform=src2.transform,
            src_crs=src2.crs,
            dst_transform=profile1["transform"],
            dst_crs=profile1["crs"],
            resampling=Resampling.bilinear,
        )
    else:
        print("⚠️ Salah satu file tidak punya CRS, skip reproject. Pastikan ukuran grid sama.")
        if data1.shape != data2.shape:
            raise ValueError("Ukuran grid berbeda dan tidak bisa disamakan tanpa CRS.")
        data2_resampled = data2

# Hitung perbedaan
diff = data1 - data2_resampled

# Tentukan range warna sama untuk File1 dan File2
combined_min = np.nanmin([data1, data2_resampled])
combined_max = np.nanmax([data1, data2_resampled])

# --- Flip supaya tidak terbalik ---
data1_plot = np.flipud(data1)
data2_plot = np.flipud(data2_resampled)
diff_plot  = np.flipud(diff)

# --- Plotting ---
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# --- Baris 1: Peta ---
# File 1
im1 = axes[0,0].imshow(data1_plot, cmap="viridis", vmin=combined_min, vmax=combined_max)
axes[0,0].set_title("Historical")
plt.colorbar(im1, ax=axes[0,0], shrink=0.7)

# File 2
im2 = axes[0,1].imshow(data2_plot, cmap="viridis", vmin=combined_min, vmax=combined_max)
axes[0,1].set_title("File 2 (Future)")
plt.colorbar(im2, ax=axes[0,1], shrink=0.7)

# Difference
im3 = axes[0,2].imshow(diff_plot, cmap="RdBu", vmin=-np.nanmax(abs(diff)), vmax=np.nanmax(abs(diff)))
axes[0,2].set_title("Difference (Historical - Future)")
plt.colorbar(im3, ax=axes[0,2], shrink=0.7)

# --- Baris 2: Histogram ---
axes[1,0].hist(data1.compressed(), bins=50, alpha=0.7, label="File 1", color="blue")
axes[1,0].hist(data2_resampled.compressed(), bins=50, alpha=0.7, label="File 2", color="green")
axes[1,0].set_title("Histogram Historical vs Future")
axes[1,0].legend()

# Histogram Difference
axes[1,1].hist(diff.compressed(), bins=50, color="red", alpha=0.7)
axes[1,1].set_title("Histogram Difference")

# Kosongkan panel terakhir
axes[1,2].axis("off")

plt.tight_layout()
plt.show()
```
