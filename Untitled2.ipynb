```python
# Read and explore GlobalGeoTree.csv, then plot

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Inline plots
%matplotlib inline
sns.set(style="whitegrid", context="notebook")

CSV_PATH = "/scratch/gito_aciar/data/Global_tree/GlobalGeoTree.csv"
assert os.path.exists(CSV_PATH), f"File not found: {CSV_PATH}"

# Read CSV (robust options)
df = pd.read_csv(
    CSV_PATH,
    low_memory=False,
)

print("=== Basic Info ===")
print(f"Path: {CSV_PATH}")
print(f"Shape: {df.shape}")
print("Columns:", list(df.columns))
print()

print("=== dtypes ===")
print(df.dtypes)
print()

display(df.head(10))

# Summary statistics
with pd.option_context('display.max_colwidth', 120):
    display(df.describe(include=[np.number]).T)

# Plotting
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in df.columns if c not in num_cols]

# 1) Histograms for numeric columns (up to first 12)
if num_cols:
    n_to_plot = min(12, len(num_cols))
    cols_to_plot = num_cols[:n_to_plot]
    ax = df[cols_to_plot].hist(figsize=(14, 10), bins=30, layout=(int(np.ceil(n_to_plot/3)), 3))
    plt.suptitle("Numeric distributions", y=1.02)
    plt.tight_layout()
    plt.show()
else:
    print("No numeric columns found for histograms.")

# 2) Bar plot for a small-cardinality categorical column
small_cat = None
for c in cat_cols:
    nunique = df[c].nunique(dropna=True)
    if 2 <= nunique <= 30:
        small_cat = c
        break

if small_cat is not None:
    plt.figure(figsize=(12, 5))
    df[small_cat].value_counts(dropna=False).head(20).plot(kind='bar')
    plt.title(f"Top categories for '{small_cat}'")
    plt.xlabel(small_cat)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
else:
    print("No small-cardinality categorical column found for bar plot.")

# 3) If lat/lon present, make a scatter plot
lat_candidates = ["lat", "latitude", "y", "Lat", "Latitude", "LAT"]
lon_candidates = ["lon", "lng", "longitude", "x", "Lon", "Longitude", "LON", "LNG"]

# Case-insensitive matching
lower_cols = {c.lower(): c for c in df.columns}

lat_col = next((lower_cols[c.lower()] for c in lat_candidates if c.lower() in lower_cols), None)
lon_col = next((lower_cols[c.lower()] for c in lon_candidates if c.lower() in lower_cols), None)

if lat_col and lon_col:
    # Coerce to numeric in case they are strings
    lat_values = pd.to_numeric(df[lat_col], errors='coerce')
    lon_values = pd.to_numeric(df[lon_col], errors='coerce')
    valid = lat_values.notna() & lon_values.notna()

    if valid.sum() > 0:
        plt.figure(figsize=(8, 6))
        plt.scatter(lon_values[valid], lat_values[valid], s=5, alpha=0.5)
        plt.title(f"Geographic scatter ({lat_col} vs {lon_col})")
        plt.xlabel(lon_col)
        plt.ylabel(lat_col)
        plt.tight_layout()
        plt.show()
    else:
        print(f"Lat/Lon columns detected ('{lat_col}', '{lon_col}') but contain no valid numeric rows.")
else:
    print("No latitude/longitude columns detected for geographic scatter.")

print("Done.")
```

```python

# Read and plot a land cover GeoTIFF
import os
from pathlib import Path
import numpy as np
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt

# Set your GeoTIFF path here. You can change this to any valid path.
TIFF_PATH = Path("/scratch/gito_aciar/sdm-toolbox/out/leptocybe-invasa/input/train/land-cover_biased_south-east-asia.tif")
# TIFF_PATH = Path("/scratch/gito_aciar/sdm-toolbox/out/leptocybe-invasa/input/train/forest-type_biased_south-east-asia.tif")
# Resolve user home (~) and make absolute for clarity
TIFF_PATH = TIFF_PATH.expanduser().resolve() if not TIFF_PATH.is_absolute() else TIFF_PATH

if not TIFF_PATH.exists():
    raise FileNotFoundError(f"File not found: {TIFF_PATH}. Please verify the path.")

with rasterio.open(TIFF_PATH) as src:
    data = src.read(1, masked=True)  # first band as masked array (nodata handled)
    bounds = src.bounds
    crs = src.crs
    transform = src.transform
    nodata = src.nodata
    dtype = src.dtypes[0]
    shape = data.shape

print("GeoTIFF info:")
print(f"  Path      : {TIFF_PATH}")
print(f"  CRS       : {crs}")
print(f"  Bounds    : {bounds}")
print(f"  Transform : {transform}")
print(f"  NoData    : {nodata}")
print(f"  DType     : {dtype}")
print(f"  Shape     : {shape} (rows, cols)")

# Choose a discrete colormap for categorical land cover data
cmap = plt.get_cmap('tab20')  # up to 20 distinct colors

# Compute unique classes (excluding masked nodata)
unique_vals = np.unique(data.compressed()) if np.ma.isMaskedArray(data) else np.unique(data)

fig, ax = plt.subplots(1, 1, figsize=(10, 6), constrained_layout=True)
show(data, transform=transform, ax=ax, cmap=cmap)
ax.set_title("Land Cover (Forest Type)  raster plot")
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

# Add a colorbar with discrete ticks when not too many classes
mappable = plt.cm.ScalarMappable(cmap=cmap)
if unique_vals.size and unique_vals.size <= 20:
    # set mappable normalization to the class range
    mappable.set_array(unique_vals)
    cbar = plt.colorbar(mappable, ax=ax, fraction=0.030, pad=0.02)
    cbar.set_ticks(unique_vals)
    cbar.set_label("Class")
else:
    # fallback generic colorbar
    mappable.set_array(data)
    cbar = plt.colorbar(mappable, ax=ax, fraction=0.030, pad=0.02)
    cbar.set_label("Value")

plt.show()

# Optional: quick value distribution as a bar chart for categorical data
if unique_vals.size and unique_vals.size <= 40:
    vals, counts = np.unique(data.compressed(), return_counts=True)
    fig2, ax2 = plt.subplots(1, 1, figsize=(10, 3), constrained_layout=True)
    ax2.bar(vals, counts, color="tab:green")
    ax2.set_title("Value counts")
    ax2.set_xlabel("Class value")
    ax2.set_ylabel("Pixel count")
    ax2.grid(True, axis='y', alpha=0.3)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import rasterio
import numpy as np
import matplotlib.patches as mpatches

# --- Dictionary kelas hutan ---
forest_dict = {
    0: ['N/K', 'Unknown', 0, 'white'],           
    1: ['ENF', 'Evergreen Needleleaf Forest', 0.8, 'red'],    
    2: ['EBF', 'Evergreen Broadleaf Forest', 1.0, 'orange'],  
    3: ['DNF', 'Deciduous Needleleaf Forest', 0.2, 'green'],  
    4: ['DBF', 'Deciduous Broadleaf Forest', 0.4, 'blue'],    
    5: ['MF', 'Mixed Forest', 0.5, 'purple']                  
}

# # --- 1. Baca file CSV ---
# csv_file = "data.csv"
# df = pd.read_csv(csv_file)

# # --- 2. Baca file GeoTIFF ---
# tif_file = "data.tif"
with rasterio.open(tif_file) as src:
    bounds = src.bounds

    # Filter data CSV agar hanya dalam bbox raster
    df_filtered = df[
        (df["longitude"] >= bounds.left) & (df["longitude"] <= bounds.right) &
        (df["latitude"]  >= bounds.bottom) & (df["latitude"]  <= bounds.top)
    ]

    # Konversi kolom level0 ke kategori numerik
    df_filtered["level0_cat"] = df_filtered["level0"].astype("category")
    df_filtered["level0_code"] = df_filtered["level0_cat"].cat.codes
    categories = df_filtered["level0_cat"].cat.categories

    # Ambil array raster
    raster_data = src.read(1)  # band pertama
    extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

    # Buat colormap dari forest_dict
    cmap_colors = [forest_dict[k][3] for k in sorted(forest_dict.keys())]
    cmap_bounds = sorted(forest_dict.keys())

    # Buat peta raster dengan warna kategori
    fig, ax = plt.subplots(figsize=(10, 8))
    im = ax.imshow(raster_data, cmap=plt.matplotlib.colors.ListedColormap(cmap_colors),
                   vmin=min(cmap_bounds), vmax=max(cmap_bounds),
                   extent=extent, origin="upper")

    # Buat legenda raster pakai forest_dict
    raster_patches = [
        mpatches.Patch(color=forest_dict[k][3], label=forest_dict[k][1])
        for k in forest_dict.keys()
    ]
    raster_legend = ax.legend(handles=raster_patches, title="Forest Classes (Raster)",
                              bbox_to_anchor=(1.05, 1), loc="upper left")

    # Tambah legenda raster ke axis
    ax.add_artist(raster_legend)

    # Plot titik CSV
    scatter = ax.scatter(
        df_filtered["longitude"], df_filtered["latitude"],
        c=df_filtered["level0_code"],
        cmap="tab20", s=30, edgecolor="black"
    )

    # Legenda untuk kategori CSV
    csv_handles = []
    for code, cat in enumerate(categories):
        csv_handles.append(
            plt.Line2D([0], [0], marker="o", color="w",
                       markerfacecolor=plt.cm.tab20(code / len(categories)),
                       markersize=8, label=cat)
        )

    ax.legend(handles=csv_handles, title="Level0 (CSV)", bbox_to_anchor=(1.05, 0.5), loc="center left")

    ax.set_title("GeoTIFF (Forest Classes) + Titik CSV", fontsize=14)
    plt.show()
```
